```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
```

```python
# Load Dataset
data = pd.read_csv('./data/train_1.csv').fillna(0)
page = data['Page']
data.head()
```

```python
# parse Page column
split_page = data["Page"].str.rsplit("_", n = 3, expand = True)
data.insert(loc=0, column='name', value=split_page[0])
data.insert(loc=1, column='project', value=split_page[1])
data.insert(loc=2, column='access', value=split_page[2])
data.insert(loc=3, column='agent', value=split_page[3])
data = data.drop(["Page"], axis=1)

split_language = data["project"].str.split(".", n = 1, expand = True)
data.insert(loc=1, column='language', value=split_language[0])
data = data.drop(["project"], axis=1)

data.head()
```

```python
# helper functions
def sumAllRows(df, row_name):
    columns = list(df.columns.values)
    sum_rows = pd.DataFrame(columns=columns)
    for column in columns:
        if all(type(x) is not str for x in df[column]):
            sum_rows.at[row_name, column] = df[column].sum()
    return sum_rows
```

```python
# get dataset sum by language
columns = list(data.columns.values)
use_by_language = pd.DataFrame(columns=columns)
languages = data.language.unique()
for language in languages:
    usage = data[data.language == language]
    use_by_language = use_by_language.append(sumAllRows(usage, language))

display(use_by_language.head())

# get dataset sum by access
columns = list(data.columns.values)
use_by_access = pd.DataFrame(columns=columns)
accesses = data.access.unique()
for access in accesses:
    usage = data[data.access == access]
    use_by_access = use_by_access.append(sumAllRows(usage, access))

display(use_by_access.head())

# get dataset sum by agent
columns = list(data.columns.values)
use_by_agent = pd.DataFrame(columns=columns)
agents = data.agent.unique()
for agent in agents:
    usage = data[data.agent == agent]
    use_by_agent = use_by_agent.append(sumAllRows(usage, agent))

display(use_by_agent.head())
```
